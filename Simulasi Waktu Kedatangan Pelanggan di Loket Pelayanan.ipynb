```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameter lambda
lambda_val = 1/5  # Rata-rata pelanggan datang tiap 5 menit
n = 30  # Jumlah iterasi

# Simulasi waktu antar kedatangan
inter_arrival_times = np.random.exponential(scale=1/lambda_val, size=n)

# Visualisasi
plt.figure(figsize=(10, 5))
sns.histplot(inter_arrival_times, bins=10, kde=True, color='skyblue')
plt.title('Simulasi Waktu Antar Kedatangan Pelanggan')
plt.xlabel('Waktu (menit)')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()

# Nilai rata-rata simulasi
mean_simulasi = np.mean(inter_arrival_times)
print(f'Rata-rata waktu antar kedatangan (simulasi): {mean_simulasi:.2f} menit')
print(f'Rata-rata teoritis eksponensial: {1/lambda_val:.2f} menit')
```
